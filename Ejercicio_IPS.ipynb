```python
class IPS:
    def __init__(self):
        self.df = pd.read_excel("FilesDS/ips.xlsx")
        print("IPS Cargado")

    def rangoPrueba(self, rango, x, y):
        #print("Rango de prueba", rango, x, y)
        #print(self.df[(self.df[rango]>=x) & (self.df[rango]<=y)])
        #loop rango de prueba lambda
        print(self.df[(self.df[rango]>=x) & (self.df[rango]<=y)].apply(lambda x: x, axis=1))

i = IPS()

i.rangoPrueba("SimceLectura20198v",270,350)


```
