4.- Se sabe que el ingreso per cápida equivalente es igual a
      IPE = (Ingreso total grupo familiar ...

      Cree una función que reciba como parámetro un folio de hogar y devuelva el IPE correspondiende
      Considere que el ingreso de todo el grupo familiar se encuentra en el campo ING_TOTAL_HOG_HD y que la información se toma del jefe de hogar


```python
def ipe(folio):
    return df.loc[df['FOLIO'] == folio, 'ING_TOTAL_HOG_HD'].values[0]

print('El IPE del folio 00001-1 es', ipe('00001-1'))
```

6.- Una clasificación de grupos socioeconómicos por familia es la siguiente
     IFE entre 0 y 175000 GS1

     Construya una función que reciba como parámetro un folio de hogar e indique en qué GS está

```python
#clasificar por ING_TOTAL_HOG_HD entre 0 y 175000 

def clasificar_ingreso(folio):
    print(df.loc[df['FOLIO'] == folio, 'ING_TOTAL_HOG_HD'])
    ingreso = df.loc[df['FOLIO'] == folio, 'ING_TOTAL_HOG_HD'].values[0]
    if ingreso <= 175000:
        return 'GS1'
    elif ingreso <= 350000:
        return 'GS2'
    else:
        return 'GS3'
    
print('El folio 00001-1 pertenece al grupo socioeconómico', clasificar_ingreso('00001-1'))
```
