¿Cuál fue el porcentaje de sobrevivencia en el Titanic? Calcula y muestra este valor.
x

```python
#Ingrese aquí su solución
survived = df[df['Survived']==1]
survived_percent = len(survived)/len(df)*100
print('Porcentaje de sobrevivientes: ',survived_percent.__round__(2))
plt.pie([survived_percent,100-survived_percent],labels=['Sobrevivientes','Muertos'],autopct='%1.1f%%',colors=['green','red'])
```
