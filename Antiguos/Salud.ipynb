```python
#Muestre los datos de Puente Alto, La Pintana y La Florida
comunas = ['Puente Alto', 'La Florida', 'La Pintana']
df_selected = df_salud.loc[df_salud['COMUNA'].isin(comunas)]
print(df_selected)
```

```python
#Muestre las comuna que tiene mayor poblacion
df_salud[['COMUNA', 'POB']][df_salud.POB==df_salud['POB'].max()]
```

```python
#Muestre la comuna que tiene mayor mortalidad
df_salud[['COMUNA', 'TMORT']][df_salud.TMORT==df_salud['TMORT'].max()]
```

```python
#Muestre la comuna que tienen menos areas verdes
df_salud[['COMUNA', 'AVERDE']][df_salud.AVERDE==df_salud['AVERDE'].min()]
```
