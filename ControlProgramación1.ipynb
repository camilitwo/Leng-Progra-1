```python
class Fraccion:
    def __init__(self, numerador, denominador):
        self.numerador = numerador
        self.denominador = Utilitario.validaDenominador(denominador)

    def __str__(self):
        return f'{self.numerador}/{self.denominador}'
    
    def simplificar(self):
        mcd = Utilitario.minimoComunMultiplo(self.numerador, self.denominador)
        self.numerador = self.numerador // mcd
        self.denominador = self.denominador // mcd
    def imprimir(self):
        Utilitario.fraccionMixta(self.numerador, self.denominador)

class HartasFracciones:
    def __init__(self):
        self.fracciones = []
    
    def agregarFraccion(self, fraccion):
        self.fracciones.append(fraccion)

    def sumatoriaDeFracciones(self):
        numerador = 0
        denominador = 1
        for fraccion in self.fracciones:
            numerador = numerador * fraccion.denominador + fraccion.numerador * denominador
            denominador = denominador * fraccion.denominador
        fraccion = Fraccion(numerador, denominador)
        fraccion.simplificar()
        fraccion.imprimir()

    def multiplicatoriaDeFracciones(self):
        numerador = 1
        denominador = 1
        for fraccion in self.fracciones:
            numerador = numerador * fraccion.numerador
            denominador = denominador * fraccion.denominador
        fraccion = Fraccion(numerador, denominador)
        fraccion.simplificar()
        fraccion.imprimir()
        
    def imprimirFracciones(self):
        for fraccion in self.fracciones:
            fraccion.imprimir()
    
class Utilitario:
    def minimoComunMultiplo(a, b):
        while b != 0:
            a, b = b, a % b
        return a
    
    def fraccionMixta(numerador, denominador):
        if numerador > denominador:
            parteEntera = numerador // denominador
            numerador = numerador % denominador
            fraccion = Fraccion(numerador, denominador)
            if numerador > 0:
                fraccion = Fraccion(numerador, denominador)
                fraccion.simplificar()
                print(f'{parteEntera}' +'E '+ f'{fraccion}')
            else:
                print(f'{parteEntera}')
        elif numerador == denominador:
            print('1E')
        else:
            fraccion = Fraccion(numerador, denominador)
            fraccion.simplificar()
            print(fraccion)

    def validaDenominador(denominador):
        if denominador == 0:
            raise LookupError('No se puede dividir entre 0').with_traceback(None)
        return denominador

def cualEsMayor(fraccion1, fraccion2):
    if fraccion1.numerador / fraccion1.denominador > fraccion2.numerador / fraccion2.denominador:
        print(f'{fraccion1} es mayor que {fraccion2}')
    elif fraccion1.numerador / fraccion1.denominador == fraccion2.numerador / fraccion2.denominador:
        print(f'{fraccion1} es igual que {fraccion2}')
    else:
        print(f'{fraccion2} es mayor que {fraccion1}')



fraccion1 = Fraccion(10, -12)
fraccion2 = Fraccion(2, -2)
fraccion3 = Fraccion(3, 9)

hartasFracciones = HartasFracciones()
hartasFracciones.agregarFraccion(fraccion1)
hartasFracciones.agregarFraccion(fraccion2)
hartasFracciones.agregarFraccion(fraccion3)

hartasFracciones.imprimirFracciones()

print('\n\nSumatoria de fracciones')
hartasFracciones.sumatoriaDeFracciones()

print('\n\nMultiplicatoria de fracciones')
hartasFracciones.multiplicatoriaDeFracciones()

print('\n\nCual es mayor')
cualEsMayor(fraccion2, fraccion3)


#Validacion de denominador 0
fraccion1 = Fraccion(10, 0)


```
